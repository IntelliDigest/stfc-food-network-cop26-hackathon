## Example API requests to data.gov.in

```python
import requests
import pandas as pd
from io import BytesIO

def get_as_dataframe(url, api_params):
    r = requests.api.get(url, api_params)
    return pd.read_csv(BytesIO(r.content))

# address of resource
url = 'https://api.data.gov.in/resource/af4ed290-ed4f-40e1-a8b8-a4440e57a9ed'

# additional params
api_params = {
    'api-key': '579b464db66ec23bdd000001cdd3946e44ce4aad7209ff7b23ac571b', # change this for the one generated by your user account
    'format': 'csv' # The output format for the data (XML or CSV)
}

df = get_as_dataframe(url, api_params)
df.head()
```
